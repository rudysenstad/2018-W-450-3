```python
from pymongo import MongoClient
```

```python
x = open('mongo_access.csv', 'r').readline().split(",")
ip_address, port, sample_size = (x[0], int(x[1]), int(x[2]))
```

```python
mongo_client = MongoClient(ip_address, port)
database_reference = mongo_client.twitter
collection_reference = database_reference.tweets
```

```python
cursor = collection_reference.find()
print("Total Tweet Count in DB: {:,}".format(cursor.count()))
```

```python
cursor_sampl = collection_reference.aggregate([{'$sample': {'size': 5}}])
tweets_samples = list(cursor_sampl)
print("Number of Tweets: {:,}".format(len(tweets_samples)))
```
